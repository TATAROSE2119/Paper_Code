```python
import numpy as np
import matplotlib.pyplot as plt

# 生成两组数据：一个密集的样本空间和一个稀疏的样本空间
np.random.seed(0)

# 密集的样本空间
dense_samples = np.random.normal(loc=0, scale=3, size=(10, 2))

# 稀疏的样本空间
sparse_samples = np.random.normal(loc=0, scale=10, size=(10, 2))

# 绘制两个样本空间的图
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 密集样本空间
axes[0].scatter(dense_samples[:, 0], dense_samples[:, 1], color='blue', edgecolor='k', alpha=0.7)
axes[0].set_title('dense_samples')
axes[0].set_xlim(-20, 20)
axes[0].set_ylim(-20, 20)

# 稀疏样本空间
axes[1].scatter(sparse_samples[:, 0], sparse_samples[:, 1], color='red', edgecolor='k', alpha=0.7)
axes[1].set_title('sparse_samples')
axes[1].set_xlim(-20, 20)
axes[1].set_ylim(-20, 20)

# 显示图像
plt.show()
```
